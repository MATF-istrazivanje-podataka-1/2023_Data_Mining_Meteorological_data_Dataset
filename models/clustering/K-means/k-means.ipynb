```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.cluster import BisectingKMeans
from fcmeans import FCM
import random
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
```

```python
data = pd.read_csv("~/ip_timeseries/train_timeseries.csv")
```

```python
# imamo jedan object: date
data.info()
```

```python
# sadr≈æi NaN vrednosti u score
data.isna().any()
```

```python
# izbacujemo sve vrste sa NaN za score
# score se meri samo jednom u nedelju dana
# resetujemo indeks da bi poƒçinjao od 0
data = data[data['score'].notnull()].reset_index(drop=True)
data
```

```python
# formatiranje
# formatiranje datuma
data['year'] = pd.DatetimeIndex(data['date']).year
data['month'] = pd.DatetimeIndex(data['date']).month
data['day'] = pd.DatetimeIndex(data['date']).day
# i izbacivanje date po≈°to postaje redundantan
data = data.drop('date', axis = 1).reset_index(drop=True)
# formatiranje score-a
data['score'] = data['score'].round().astype(int)
```

```python
# feature selection i feature extraction
# T2M_RANGE, WS10M_RANGE i WS50M_RANGE su redundantne i
# mogu da se izbace po≈°to imamo i max i min za ove vrednosti
data = data.drop('T2M_RANGE', axis = 1)
data = data.drop('WS10M_RANGE', axis = 1)
data = data.drop('WS50M_RANGE', axis = 1)
```

```python
for i in data.columns:
    if i == 'fips' or i == 'day' or i == 'month' or i == 'year' or i == 'score':
        continue
    data = data[(data[f'{i}'] <= data[f'{i}'].mean() + 3 * data[f'{i}'].std())
              & (data[f'{i}'] >= data[f'{i}'].mean() - 3 * data[f'{i}'].std())]
# provera da li smo dobro izbacili outlier-e
data.shape
```

```python
X = data.drop('score', axis = 1)
X = X.drop('fips', axis = 1)
y = data['score']
```

```python
scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)
X = pd.DataFrame(data=X_scaled, columns=X.columns)
X
```

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)
```

```python
X_pca.shape
```

```python
#uvodimo parametre i biramo deo podataka nasumiƒçno (1/10)
k = range(2, 10)
palette = sns.color_palette()
X_pca_small = X_pca[np.random.randint(X_pca.shape[0], size = 250000), :]
X_pca_small.shape
```

```python
Kmean_calinski = []
Kmean_davies = []
```

```python
sse = []
silhouettes = []

fig, axs = plt.subplots(4, 2, figsize=(12, 25))
fig.suptitle("K-sredina", y = 1, fontsize = 30)
for i, ax in zip(k, axs.flatten()):
    Kmean = KMeans(n_clusters = i, n_init = 'auto', algorithm='elkan')
    Kmean.fit(X_pca_small)
    sse.append(Kmean.inertia_)
    silhouettes.append(silhouette_score(X_pca_small, Kmean.labels_))
    
    Kmean_calinski.append(calinski_harabasz_score(X_pca_small, Kmean.labels_))
    Kmean_davies.append(davies_bouldin_score(X_pca_small, Kmean.labels_))
    
    ax.scatter(X_pca_small[:, 0], X_pca_small[:, 1], s = 50, c = Kmean.labels_)
    ax.set_title(f"{i} klastera")
    ax.set_xlabel("Komponenta 1")
    ax.set_ylabel("Komponenta 2")
    for j in range (i):
        ax.scatter(Kmean.cluster_centers_[j][0], Kmean.cluster_centers_[j][1], s=200, color=palette[j], marker='s')
plt.tight_layout()
plt.show()
```

```python
plt.plot(k, sse)
plt.xlabel('Broj klastera')
plt.ylabel('Skor sse')
```

```python
plt.plot(k, silhouettes)
plt.xlabel('Broj klastera')
plt.ylabel('Skor silueta')
```

```python
fcm_calinski = []
fcm_davies = []
```

```python
sse_fcm = []
silhouettes_fcm = []

fig, axs = plt.subplots(4, 2, figsize=(12, 25))
fig.suptitle('Fuzzy klasterovanje', y = 1, fontsize = 30)
for i, ax in zip(k, axs.flatten()):
    fcm = FCM(n_clusters = i)
    fcm.fit(X_pca_small)
    
    centers = fcm.centers
    labels = fcm.predict(X_pca_small)
    
    fcm_calinski.append(calinski_harabasz_score(X_pca_small, labels))
    fcm_davies.append(davies_bouldin_score(X_pca_small, labels))
    
    ax.scatter(X_pca_small[:, 0], X_pca_small[:, 1], s = 50, c = labels)
    ax.set_title(f"{i} klastera")
    ax.set_xlabel("Komponenta 1")
    ax.set_ylabel("Komponenta 2")
    for j in range (i):
        ax.scatter(centers[j][0], centers[j][1], s = 200, color = palette[j], marker = 's')
plt.tight_layout()
plt.show()
```

```python
BKmean_calinski = []
BKmean_davies = []
```

```python
sse_bisecting = []
silhouettes_bisecting = []

fig, axs = plt.subplots(4, 2, figsize=(12, 25))
fig.suptitle("Bisektivno k-sredina", y = 1, fontsize = 30)
for i, ax in zip(k, axs.flatten()):
    BKmean = BisectingKMeans(n_clusters = i, algorithm = 'elkan')
    BKmean.fit(X_pca_small)
    sse_bisecting.append(BKmean.inertia_)
    silhouettes_bisecting.append(silhouette_score(X_pca_small, BKmean.labels_))
    
    BKmean_calinski.append(calinski_harabasz_score(X_pca_small, BKmean.labels_))
    BKmean_davies.append(davies_bouldin_score(X_pca_small, BKmean.labels_))
    
    ax.scatter(X_pca_small[:, 0], X_pca_small[:, 1], s = 50, c = BKmean.labels_)
    ax.set_title(f"{i} klastera")
    ax.set_xlabel("Komponenta 1")
    ax.set_ylabel("Komponenta 2")
    for j in range (i):
        ax.scatter(BKmean.cluster_centers_[j][0], BKmean.cluster_centers_[j][1], s=200, color=palette[j], marker='s')
plt.tight_layout()
plt.show()
```

```python
plt.plot(k, sse_bisecting)
plt.xlabel('Broj klastera')
plt.ylabel('Skor sse')
```

```python
plt.plot(k, silhouettes_bisecting)
plt.xlabel('Broj klastera')
plt.ylabel('Skor silueta')
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 12))
fig.suptitle("Calinski i Davies mera za modele klasterovanja", y = 1, fontsize = 30)

axs = axs.flatten()
axs[0].plot(k, Kmean_calinski)
axs[0].set_xlabel('Broj klastera')
axs[0].set_ylabel('Calinski-Harabasz skor')
axs[0].set_title("Kmean", x = 1.1, y = 1.1, size = 20)
axs[1].plot(k, Kmean_davies)
axs[1].set_xlabel('Broj klastera')
axs[1].set_ylabel('Davies-Bouldin skor')

axs[2].plot(k, fcm_calinski)
axs[2].set_xlabel('Broj klastera')
axs[2].set_ylabel('Calinski-Harabasz skor')
axs[2].set_title("FCM", x = 1.1, y = 1.1, size = 20)
axs[3].plot(k, fcm_davies)
axs[3].set_xlabel('Broj klastera')
axs[3].set_ylabel('Davies-Bouldin skor')

axs[4].plot(k, BKmean_calinski)
axs[4].set_xlabel('Broj klastera')
axs[4].set_ylabel('Calinski-Harabasz skor')
axs[4].set_title("BKmean", x = 1.1, y = 1.1, size = 20)
axs[5].plot(k, BKmean_davies)
axs[5].set_xlabel('Broj klastera')
axs[5].set_ylabel('Davies-Bouldin skor')

plt.tight_layout()
plt.show()
```
