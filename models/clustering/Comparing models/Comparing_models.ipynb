```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
```

```python
%store -r Kmean_calinski
%store -r Kmean_davies
%store -r Kmean_calinski_pca
%store -r Kmean_davies_pca
%store -r BKmean_calinski
%store -r BKmean_davies
%store -r BKmean_calinski_pca
%store -r BKmean_davies_pca
%store -r fcm_calinski_pca
%store -r fcm_davies_pca
%store -r birch_calinski_pca
%store -r birch_davies_pca

%store -r sse
%store -r silhouettes
%store -r sse_pca
%store -r silhouettes_pca
%store -r sse_bisecting
%store -r silhouettes_bisecting
%store -r sse_bisecting_pca
%store -r silhouettes_bisecting_pca
```

```python
k = range(2, 10)
```

```python
fig, axs = plt.subplots(4, 2, figsize=(12, 16))
fig.suptitle("SSE i skor silueta za KMeans i BKmeans", y = 1, fontsize = 30)

axs = axs.flatten()
axs[0].plot(k, sse)
axs[0].set_xlabel('Broj klastera')
axs[0].set_ylabel('Skor sse')
axs[0].set_title("Kmean", x = 1.1, y = 1.1, size = 20)
axs[1].plot(k, silhouettes)
axs[1].set_xlabel('Broj klastera')
axs[1].set_ylabel('Skor silueta')

axs[2].plot(k, sse_pca)
axs[2].set_xlabel('Broj klastera')
axs[2].set_ylabel('Skor sse')
axs[2].set_title("Kmean sa PCA", x = 1.1, y = 1.1, size = 20)
axs[3].plot(k, silhouettes_pca)
axs[3].set_xlabel('Broj klastera')
axs[3].set_ylabel('Skor silueta')

axs[4].plot(k, sse_bisecting)
axs[4].set_xlabel('Broj klastera')
axs[4].set_ylabel('Skor bisektiranog sse')
axs[4].set_title("BKmean", x = 1.1, y = 1.1, size = 20)
axs[5].plot(k, silhouettes_bisecting)
axs[5].set_xlabel('Broj klastera')
axs[5].set_ylabel('Skor bisektiranih silueta')

axs[6].plot(k, sse_bisecting_pca)
axs[6].set_xlabel('Broj klastera')
axs[6].set_ylabel('Skor bisektiranog sse')
axs[6].set_title("BKmean sa PCA", x = 1.1, y = 1.1, size = 20)
axs[7].plot(k, silhouettes_bisecting_pca)
axs[7].set_xlabel('Broj klastera')
axs[7].set_ylabel('Skor bisektiranih silueta')

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(6, 2, figsize=(12, 24))
fig.suptitle("Calinski i Davies mera za razne modele klasterovanja", y = 1, fontsize = 30)

axs = axs.flatten()
axs[0].plot(k, Kmean_calinski)
axs[0].set_xlabel('Broj klastera')
axs[0].set_ylabel('Calinski-Harabasz skor')
axs[0].set_title("Kmean", x = 1.1, y = 1.1, size = 20)
axs[1].plot(k, Kmean_davies)
axs[1].set_xlabel('Broj klastera')
axs[1].set_ylabel('Davies-Bouldin skor')

axs[2].plot(k, Kmean_calinski_pca)
axs[2].set_xlabel('Broj klastera')
axs[2].set_ylabel('Calinski-Harabasz skor')
axs[2].set_title("Kmean sa PCA", x = 1.1, y = 1.1, size = 20)
axs[3].plot(k, Kmean_davies_pca)
axs[3].set_xlabel('Broj klastera')
axs[3].set_ylabel('Davies-Bouldin skor')

axs[4].plot(k, fcm_calinski_pca)
axs[4].set_xlabel('Broj klastera')
axs[4].set_ylabel('Calinski-Harabasz skor')
axs[4].set_title("FCM sa PCA", x = 1.1, y = 1.1, size = 20)
axs[5].plot(k, fcm_davies_pca)
axs[5].set_xlabel('Broj klastera')
axs[5].set_ylabel('Davies-Bouldin skor')

axs[6].plot(k, BKmean_calinski)
axs[6].set_xlabel('Broj klastera')
axs[6].set_ylabel('Calinski-Harabasz skor')
axs[6].set_title("BKmean", x = 1.1, y = 1.1, size = 20)
axs[7].plot(k, BKmean_davies)
axs[7].set_xlabel('Broj klastera')
axs[7].set_ylabel('Davies-Bouldin skor')

axs[8].plot(k, BKmean_calinski_pca)
axs[8].set_xlabel('Broj klastera')
axs[8].set_ylabel('Calinski-Harabasz skor')
axs[8].set_title("BKmean sa PCA", x = 1.1, y = 1.1, size = 20)
axs[9].plot(k, BKmean_davies_pca)
axs[9].set_xlabel('Broj klastera')
axs[9].set_ylabel('Davies-Bouldin skor')

axs[10].plot(k, birch_calinski_pca)
axs[10].set_xlabel('Broj klastera')
axs[10].set_ylabel('Calinski-Harabasz skor')
axs[10].set_title("BIRCH sa PCA", x = 1.1, y = 1.1, size = 20)
axs[11].plot(k, birch_davies_pca)
axs[11].set_xlabel('Broj klastera')
axs[11].set_ylabel('Davies-Bouldin skor')

plt.tight_layout()
plt.show()
```
