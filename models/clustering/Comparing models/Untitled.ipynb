```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
```

```python
%store -r Kmean_calinski
%store -r Kmean_davies
%store -r BKmean_calinski
%store -r BKmean_davies
%store -r fcm_calinski
%store -r fcm_davies
%store -r birch_calinski
%store -r birch_davies

%store -r sse
%store -r silhouettes
%store -r sse_bisecting
%store -r silhouettes_bisecting
```

```python
k = range(2, 10)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
fig.suptitle("SSE i skor silueta za KMeans i BKmeans", y = 1, fontsize = 30)

axs = axs.flatten()
axs[0].plot(k, sse)
axs[0].set_xlabel('Broj klastera')
axs[0].set_ylabel('Skor sse')
axs[0].set_title("Kmean", x = 1.1, y = 1.1, size = 20)
axs[1].plot(k, silhouettes)
axs[1].set_xlabel('Broj klastera')
axs[1].set_ylabel('Skor silueta')

axs[2].plot(k, sse_bisecting)
axs[2].set_xlabel('Broj klastera')
axs[2].set_ylabel('Skor bisektiranog sse')
axs[2].set_title("BKmean", x = 1.1, y = 1.1, size = 20)
axs[3].plot(k, silhouettes_bisecting)
axs[3].set_xlabel('Broj klastera')
axs[3].set_ylabel('Skor bisektiranih silueta')

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(4, 2, figsize=(12, 20))
fig.suptitle("Calinski i Davies mera za razne modele klasterovanja", y = 1, fontsize = 30)

axs = axs.flatten()
axs[0].plot(k, Kmean_calinski)
axs[0].set_xlabel('Broj klastera')
axs[0].set_ylabel('Calinski-Harabasz skor')
axs[0].set_title("Kmean", x = 1.1, y = 1.1, size = 20)
axs[1].plot(k, Kmean_davies)
axs[1].set_xlabel('Broj klastera')
axs[1].set_ylabel('Davies-Bouldin skor')

axs[2].plot(k, fcm_calinski)
axs[2].set_xlabel('Broj klastera')
axs[2].set_ylabel('Calinski-Harabasz skor')
axs[2].set_title("FCM", x = 1.1, y = 1.1, size = 20)
axs[3].plot(k, fcm_davies)
axs[3].set_xlabel('Broj klastera')
axs[3].set_ylabel('Davies-Bouldin skor')

axs[4].plot(k, BKmean_calinski)
axs[4].set_xlabel('Broj klastera')
axs[4].set_ylabel('Calinski-Harabasz skor')
axs[4].set_title("BKmean", x = 1.1, y = 1.1, size = 20)
axs[5].plot(k, BKmean_davies)
axs[5].set_xlabel('Broj klastera')
axs[5].set_ylabel('Davies-Bouldin skor')

axs[6].plot(k, birch_calinski)
axs[6].set_xlabel('Broj klastera')
axs[6].set_ylabel('Calinski-Harabasz skor')
axs[6].set_title("BIRCH", x = 1.1, y = 1.1, size = 20)
axs[7].plot(k, birch_davies)
axs[7].set_xlabel('Broj klastera')
axs[7].set_ylabel('Davies-Bouldin skor')

plt.tight_layout()
plt.show()
```
